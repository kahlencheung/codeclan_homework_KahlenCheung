### Question 1

Read in all 4 credit card transaction datasets and clean up the column names.

```python
import pandas as pd
from janitor import clean_names
```

```python
card = pd.read_csv("data/CardBase.csv")
customer = pd.read_csv("data/CustomerBase.csv")
fraud = pd.read_csv("data/FraudBase.csv")
transaction = pd.read_csv("data/TransactionBase.csv")
```

```python
card = card.dropna()
customer = customer.dropna()
fraud = fraud.dropna()
transaction = transaction.dropna()
```

```python
card.columns = map(str.lower, card.columns)
```

```python
customer.columns = map(str.lower, customer.columns)
```

```python
fraud.columns = map(str.lower, fraud.columns)
```

```python
transaction.columns = map(str.lower, transaction.columns)
```
