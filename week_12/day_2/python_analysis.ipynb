3. Select the variables Beverage_category, Beverage, Beverage prep and Calories from the drinks_content data frame, and assign the selected columns to a new data frame called drinks. Check if there are any missing values anywhere in the data, and drop any rows containing them if there are.

```python
drinks = drinks_content.loc[:,['Beverage_category', 'Beverage', 'Beverage_prep', 'Calories']].copy()
drinks
```

```python
drinks = drinks.dropna()
drinks
```

4. In your drinks dataframe create a new variable (column) called calorie_diff, which stores the difference between 135 calories (135 calories = 10 cubes of sugar!) and the calories in each drink. (Hint: you’ll want to subtract the calories from 135 to see which drink have more or less than 10 cups of sugar).

```python
drinks['calorie_diff'] = drinks['Calories'] - 135
drinks
```

5. Summarise the mean number of calories in each beverage category.

```python
drinks.groupby('Beverage_category')['Calories'].mean()
```

6. Harder - Which Beverage_prep type contains more than the average calories of all drinks? (Hint: to answer this, you’ll have to first figure out what the average calories across all drinks are, and then use that as a filter for the grouped Beverage_prep data).

```python
drinks.loc[drinks['Calories'] > drinks['Calories'].mean()].groupby('Beverage_category').count()['Calories']
```

7. Which is the best type of coffee drink to get if you’re worried about consuming too many calories? (Hint: you can use pd.unique() on a column to get all unique values/levels of that column! Also remember that you can use isin() to subset with a list)

```python
drinks.loc[drinks['Beverage'].isin(['Coffee'])].groupby('Beverage_category').head().sort_values(by = 'Calories')
```

8. Filter the dataset drinks so we only take "Classic Espresso Drinks", and save this in a new data frame called espresso_drinks.

```python
espresso_drinks = drinks.loc[drinks['Beverage_category'] == 'Classic Espresso Drinks'].copy()
```

```python
espresso_drinks
```

9. Group your espresso_drinks data frame by the type of Beverage prep, and then find out the mean calories for each drink group.

```python
espresso_drinks.groupby('Beverage_prep').mean()['Calories']
```

10. Get the same grouped mean Calories values as above, but this time sorted in descending order. (Hint: so you're able to sort the resulting groupby by the 'Calories' column you will want to ensure you get back a dataframe and not a series)

```python
espresso_drinks.groupby('Beverage_prep').mean()['Calories'].to_frame().sort_values(by = 'Calories')
```
